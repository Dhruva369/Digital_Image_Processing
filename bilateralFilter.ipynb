```python
# program using OpenCV in Python to implement Bilateral filter
# removal of salt and pepper noise while preserving edges
#--------------------------------

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/lena_salt_pepper.jpg")

bimg1 = cv2.bilateralFilter(img, 9, 75, 75)
bimg2 = cv2.bilateralFilter(img, 9, 175, 175)
bimg3 = cv2.bilateralFilter(img, 15, 175, 175)

plt.subplot(221),plt.imshow(img),plt.title('Original'),plt.axis('off')
plt.subplot(222),plt.imshow(bimg1),plt.title('Bilateral 9-75'),plt.axis('off')
plt.subplot(223),plt.imshow(bimg2),plt.title('Bilateral 9-175'),plt.axis('off')
plt.subplot(224),plt.imshow(bimg3),plt.title('Bilateral 15-175'),plt.axis('off')

plt.show()
```
