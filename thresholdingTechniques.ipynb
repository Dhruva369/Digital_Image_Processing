```python
# program using OpenCV in Python  to convert a colored image to binary image
#by applying different thresholding techniques using matplotlib
#----------------------------------

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/lena_c.jpg")

rgbimg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

ret,thresh1 = cv2.threshold(gray,120,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(gray,120,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(gray,120,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(gray,120,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(gray,120,255,cv2.THRESH_TOZERO_INV)

plt.subplot(332),plt.imshow(rgbimg,cmap='gray'),plt.title('Original'),plt.axis('off')
plt.subplot(334),plt.imshow(gray,cmap='gray'),plt.title('Grayscale'),plt.axis('off')
plt.subplot(335),plt.imshow(thresh1,cmap='gray'),plt.title('Binary'),plt.axis('off')
plt.subplot(336),plt.imshow(thresh2,cmap='gray'),plt.title('Binary Inverse'),plt.axis('off')
plt.subplot(337),plt.imshow(thresh3,cmap='gray'),plt.title('Truncated'),plt.axis('off')
plt.subplot(338),plt.imshow(thresh4,cmap='gray'),plt.title('To Zero'),plt.axis('off')
plt.subplot(339),plt.imshow(thresh5,cmap='gray'),plt.title('To Zero Inverse'),plt.axis('off')

plt.show()
```
