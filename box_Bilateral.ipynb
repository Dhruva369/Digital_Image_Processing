```python
# program for comparison of Bilateral and Box filter
#------------------------------------------------------

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/lena_c.jpg")

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

biImg = cv2.bilateralFilter(rgb,5, 75, 75)
boxImg = cv2.blur(rgb, (5,5))

plt.subplot(131),plt.imshow(rgb),plt.title('Original Image'), plt.axis('off')
plt.subplot(132),plt.imshow(biImg),plt.title('Bilateral Blur d=5'), plt.axis('off')
plt.subplot(133),plt.imshow(boxImg),plt.title('Box Blur k=5'), plt.axis('off')

plt.show()
```
