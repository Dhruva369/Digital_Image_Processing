```python
# program using OpenCV in Python to implement Box filter
#--------------------------------

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/lena_c.jpg")
rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

blur3 = cv2.blur(rgb,(3,3))
blur5 = cv2.blur(rgb,(5,5))
blur15 = cv2.blur(rgb,(15,15))

plt.subplot(221),plt.imshow(rgb,cmap='gray'),plt.title('Original'),plt.axis('off')
plt.subplot(222),plt.imshow(blur3,cmap='gray'),plt.title('Box-3'),plt.axis('off')
plt.subplot(223),plt.imshow(blur5,cmap='gray'),plt.title('Box-5'),plt.axis('off')
plt.subplot(224),plt.imshow(blur15,cmap='gray'),plt.title('Box-15'),plt.axis('off')

plt.show()
```
