```python
# program using OpenCV in Python to convert a colored image to binary image (using plots)
#------------------------------------------------

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/lena_c.jpg")

rgbimg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

ret,thresh = cv2.threshold(gray,90,255,cv2.THRESH_BINARY)

plt.subplot(131),plt.imshow(rgbimg,cmap='gray'),plt.title('Original'),plt.axis('off')
plt.subplot(132),plt.imshow(gray,cmap='gray'),plt.title('Grayscale'),plt.axis('off')
plt.subplot(133),plt.imshow(thresh,cmap='gray'),plt.title('Binary'),plt.axis('off')

plt.show()
```
