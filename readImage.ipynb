```python
# program using OpenCV in Python to Read an image in different color
#-----------------------------------------
import cv2
import matplotlib.pyplot as plt

cimg = cv2.imread("/content/colours.png",cv2.IMREAD_COLOR)      #1
gimg = cv2.imread("/content/colours.png",cv2.IMREAD_GRAYSCALE)  #0
uimg = cv2.imread("/content/colours.png",cv2.IMREAD_UNCHANGED)  #-1

plt.subplot(131),plt.imshow(cimg,cmap='gray'),plt.title('Colour'),plt.axis('off')
plt.subplot(132),plt.imshow(gimg,cmap='gray'),plt.title('Grayscale'),plt.axis('off')
plt.subplot(133),plt.imshow(uimg,cmap='gray'),plt.title('Unchanged'),plt.axis('off')

plt.show()
```
