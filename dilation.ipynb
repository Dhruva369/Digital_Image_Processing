```python
# program for dilation of an image
#------------------------------------------------------------------

import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("/content/text.jpg")

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

ret,thresh = cv2.threshold(gray,70,255, cv2.THRESH_BINARY_INV)

kernel = np.ones((5, 5), np.uint8)
print(kernel)

dimg1=cv2.dilate(thresh,kernel,iterations=1)
dimg2=cv2.dilate(thresh,kernel,iterations=2)
dimg3=cv2.dilate(thresh,kernel,iterations=3)

plt.subplot(231),plt.imshow(image, cmap='gray'),plt.title('Original Image'), plt.axis('off')
plt.subplot(232),plt.imshow(thresh, cmap='gray'),plt.title('Thresholded image'), plt.axis('off')
plt.subplot(234),plt.imshow(dimg1, cmap='gray'),plt.title('Dilation (i=1)'), plt.axis('off')
plt.subplot(235),plt.imshow(dimg2, cmap='gray'),plt.title('Dilation (i=2)'), plt.axis('off')
plt.subplot(236),plt.imshow(dimg3, cmap='gray'),plt.title('Dilation (i=3)'), plt.axis('off')

plt.show()
```
