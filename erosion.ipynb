```python
# program for erosion of an image
#------------------------------------------------------------------

import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread("/content/text.jpg")

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

ret,thresh = cv2.threshold(gray,70,255, cv2.THRESH_BINARY_INV)

kernel = np.ones((5, 5), np.uint8)
print(kernel)

eimg1 = cv2.erode(thresh, kernel, iterations=1)
eimg2 = cv2.erode(thresh, kernel, iterations=2)
eimg3 = cv2.erode(thresh, kernel, iterations=3)

plt.subplot(231),plt.imshow(image, cmap='gray'),plt.title('Original Image'), plt.axis('off')
plt.subplot(232),plt.imshow(thresh, cmap='gray'),plt.title('Thresholded image'), plt.axis('off')
plt.subplot(234),plt.imshow(eimg1, cmap='gray'),plt.title('Erosion (i=1)'), plt.axis('off')
plt.subplot(235),plt.imshow(eimg2, cmap='gray'),plt.title('Erosion (i=2)'), plt.axis('off')
plt.subplot(236),plt.imshow(eimg3, cmap='gray'),plt.title('Erosion (i=3)'), plt.axis('off')

plt.show()
```
