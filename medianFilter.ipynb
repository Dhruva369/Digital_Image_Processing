```python
# program using OpenCV in Python to implement Median filter
# removal of salt and pepper noise
#--------------------------------

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/lena_salt_pepper.jpg")

mimg3 = cv2.medianBlur(img, 3)
mimg5 = cv2.medianBlur(img, 5)
mimg7 = cv2.medianBlur(img, 7)

plt.subplot(221),plt.imshow(img),plt.title('Original'),plt.axis('off')
plt.subplot(222),plt.imshow(mimg3),plt.title('Median 3'),plt.axis('off')
plt.subplot(223),plt.imshow(mimg5),plt.title('Median 5'),plt.axis('off')
plt.subplot(224),plt.imshow(mimg7),plt.title('Medain 7'),plt.axis('off')

plt.show()
```
