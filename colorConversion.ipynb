```python
# program using OpenCV in Python to convert color image(BGR) to 1.RGB 2.Grayscale  3.HSV
#--------------------------------------

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/lena_c.jpg")

rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

plt.subplot(221),plt.imshow(img,cmap='gray'),plt.title('Original'),plt.axis('off')
plt.subplot(222),plt.imshow(rgb,cmap='gray'),plt.title('RGB'),plt.axis('off')
plt.subplot(223),plt.imshow(gray,cmap='gray'),plt.title('Grayscale'),plt.axis('off')
plt.subplot(224),plt.imshow(hsv,cmap='gray'),plt.title('HSV'),plt.axis('off')

plt.show()
```
