```python
#program using OpenCV in Python to read an image in grayscale using CvtColor() function
#-------------------------------

import cv2
from google.colab.patches import cv2_imshow

cimg = cv2.imread("/content/colours.png")

gimg = cv2.cvtColor(cimg,cv2.COLOR_BGR2GRAY)

#cv2.imshow('Original',cimg)
cv2_imshow(cimg)
print('\n')
#cv2.imshow('Gray', gimg)
cv2_imshow(gimg)

#cv2.waitKey(0)

#cv2.destroyAllWindows()
```
