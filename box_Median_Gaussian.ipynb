```python
# program for comparison of Box, Median and Gaussian filter
#------------------------------------------------------

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/lena_c.jpg")

rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

boxImg = cv2.blur(rgb, (5,5))
medImg = cv2.medianBlur(rgb,5)
gausImg= cv2.GaussianBlur(rgb,(5,5),5)

plt.subplot(232),plt.imshow(rgb),plt.title('Original Image'), plt.axis('off')
plt.subplot(234),plt.imshow(boxImg),plt.title('Box Blur(k=5*5)'), plt.axis('off')
plt.subplot(235),plt.imshow(medImg),plt.title('Median Blur(k=5*5)'), plt.axis('off')
plt.subplot(236),plt.imshow(gausImg),plt.title('Gaussian Blur(k=5*5)'), plt.axis('off')

plt.show()
```
