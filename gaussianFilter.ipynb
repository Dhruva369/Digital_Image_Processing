```python
# program using OpenCV in Python to implement Gaussian filter
#--------------------------------

import cv2
import matplotlib.pyplot as plt

img = cv2.imread("/content/lena_c.jpg")
rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

gblur5 = cv2.GaussianBlur(rgb,(5,5),5)
gblur15 = cv2.GaussianBlur(rgb,(15,15),0)
gblur155 = cv2.GaussianBlur(rgb,(15,15),5)

plt.subplot(221),plt.imshow(rgb,cmap='gray'),plt.title('Original'),plt.axis('off')
plt.subplot(222),plt.imshow(gblur5,cmap='gray'),plt.title('Gaussian 5'),plt.axis('off')
plt.subplot(223),plt.imshow(gblur15,cmap='gray'),plt.title('Gaussian 15-0'),plt.axis('off')
plt.subplot(224),plt.imshow(gblur155,cmap='gray'),plt.title('Gaussian 15-5'),plt.axis('off')

plt.show()
```
